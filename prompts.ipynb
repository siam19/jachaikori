````python
import json
import networkx as nx
import matplotlib.pyplot as plt
from pyvis.network import Network
from IPython.display import IFrame, display, HTML

def parse_entity_data(json_data):
    """Parse entity data from JSON string or dictionary"""
    if isinstance(json_data, str):
        try:
            data = json.loads(json_data)
        except json.JSONDecodeError:
            raise ValueError("Invalid JSON string")
    else:
        data = json_data
    return data

def create_graph(data):
    """Create NetworkX graph from entity data"""
    G = nx.DiGraph()
    
    # Add nodes
    for node in data.get('nodes', []):
        # Collect attributes
        attrs = {
            'name': node.get('name', ''),
            'type': node.get('type', ''),
            'label': node.get('name', '')
        }
        # Add any additional attributes
        if 'attributes' in node:
            for k, v in node['attributes'].items():
                attrs[k] = v
        
        G.add_node(node['id'], **attrs)
    
    # Add edges
    for edge in data.get('edges', []):
        G.add_edge(
            edge['source'], 
            edge['target'], 
            type=edge.get('type', ''),
            description=edge.get('description', '')
        )
    
    return G

def visualize_interactive(G, height="800px", width="100%", filename="entity_graph.html"):
    """Create interactive visualization with movable nodes"""
    net = Network(height=height, width=width, directed=True, notebook=True)
    
    # Define colors for different node types
    color_map = {
        'person': '#add8e6',  # skyblue
        'organization': '#90ee90',  # lightgreen
        'location': '#fa8072',  # salmon
        'event': '#ffff00',  # yellow
        'product': '#a020f0',  # purple
        'group': '#ffa500'   # orange
    }
    
    # Add nodes
    for node_id in G.nodes:
        node_data = G.nodes[node_id]
        
        # Create hover info with attributes
        hover_info = f"<b>{node_data['name']}</b><br>Type: {node_data['type']}"
        if 'role' in node_data:
            hover_info += f"<br>Role: {node_data['role']}"
        for key, value in node_data.items():
            if key not in ['name', 'type', 'label', 'role'] and key != 'id':
                hover_info += f"<br>{key}: {value}"
        
        # Add node with custom styling
        net.add_node(
            node_id, 
            label=node_data['name'],  # Use name as the label
            title=hover_info,
            color=color_map.get(node_data['type'], '#d3d3d3'),
            size=25,  # Slightly larger nodes
            font={'size': 16, 'face': 'arial'},
            borderWidth=2,
            borderWidthSelected=4,
            shape='dot'  # Using dot shape for all nodes
        )
    
    # Add edges with custom styling
    for source, target, data in G.edges(data=True):
        net.add_edge(
            source, 
            target, 
            title=data.get('description', ''),
            label=data.get('type', ''),
            font={'size': 12, 'align': 'middle'},
            arrows={'to': {'enabled': True, 'type': 'arrow'}},
            color={'color': '#848484', 'highlight': '#FF0000'},
            width=2,
            selectionWidth=3
        )
    
    # Configure options for interactivity and physics
    net.set_options("""
    {
      "interaction": {
        "hover": true,
        "dragNodes": true,
        "dragView": true,
        "navigationButtons": true,
        "multiselect": true,
        "selectable": true,
        "zoomView": true
      },
      "physics": {
        "enabled": true,
        "stabilization": {
          "enabled": true,
          "iterations": 1000,
          "updateInterval": 100,
          "fit": true
        },
        "barnesHut": {
          "gravitationalConstant": -8000,
          "centralGravity": 0.1,
          "springLength": 250,
          "springConstant": 0.04,
          "damping": 0.09,
          "avoidOverlap": 0.5
        },
        "solver": "barnesHut"
      },
      "layout": {
        "randomSeed": 42,
        "improvedLayout": true
      },
      "edges": {
        "smooth": {
          "enabled": true,
          "type": "dynamic"
        }
      },
      "nodes": {
        "font": {
          "size": 16
        }
      }
    }
    """)
    
    # Save and return the network
    net.show_buttons(filter_=['physics'])  # Add physics control buttons
    net.save_graph(filename)
    return net

# Example usage with the provided JSON data:
json_string = '''
{
  "nodes": [
    {
      "id": "Person_1",
      "name": "Elon Musk",
      "type": "person",
      "attributes": {
        "role": "CEO of SpaceX",
        "location": "Washington D.C."
      }
    },
    {
      "id": "Organization_1",
      "name": "SpaceX",
      "type": "organization"
    },
    {
      "id": "Person_2",
      "name": "Jeff Bezos",
      "type": "person",
      "attributes": {
        "role": "Founder of Blue Origin"
      }
    },
    {
      "id": "Organization_2",
      "name": "Blue Origin",
      "type": "organization"
    },
    {
      "id": "Event_1",
      "name": "Artemis Moon Mission Delay",
      "type": "event",
      "attributes": {
        "location": "Moon"
      }
    },
    {
      "id": "Location_1",
      "name": "Washington D.C.",
      "type": "location"
    }
  ],
  "edges": [
    {
      "source": "Person_1",
      "target": "Organization_1",
      "type": "MEMBER_OF",
      "description": "CEO of SpaceX"
    },
    {
      "source": "Person_1",
      "target": "Person_2",
      "type": "CRITICIZED",
      "description": "criticized Jeff Bezos during a press conference"
    },
    {
      "source": "Person_1",
      "target": "Organization_2",
      "type": "ACCUSED_OF",
      "description": "accused Blue Origin of delaying NASA's Artemis Moon mission"
    },
    {
      "source": "Person_1",
      "target": "Location_1",
      "type": "LOCATED_AT",
      "description": "press conference in Washington D.C."
    },
    {
      "source": "Organization_2",
      "target": "Event_1",
      "type": "DELAYED",
      "description": "delaying Artemis Moon mission"
    }
  ]
}
'''

# Parse the data and create a graph
data = parse_entity_data(json_string)
G = create_graph(data)

# Create and display interactive visualization
network = visualize_interactive(G)

# Display the interactive graph directly in the notebook
display(IFrame("entity_graph.html", width="100%", height="800px"))

# Example of how to use with your own data:
def visualize_entity_graph(json_data, height="800px", width="100%"):
    """Helper function to easily visualize any entity data"""
    data = parse_entity_data(json_data)
    G = create_graph(data)
    network = visualize_interactive(G, height=height, width=width)
    display(IFrame("entity_graph.html", width=width, height=height))
    return network

# Usage:
# my_network = visualize_entity_graph(your_json_string)
````
